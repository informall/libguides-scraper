```python
from bs4 import BeautifulSoup
import requests
import re
from utils import *
```

```python
def go():
    memo = {}
    soups = html_parser(get_text("http://community.libguides.com/?action=0&inst_type=2"))
    links = list(filter(url_filter, soups.find_all(A)))

    for i in range(len(links)):
        print(i+1, 'of', len(links))
        link = links[i]
        try:
            memo[link.text] = {}

            site = html_parser(get_text(link[HREF] + "/az.php"))
            site_id = to_site_id(site.find_all('script'))
            site_url = link[HREF] + "/az_process.php?action=520&first=all" + ("&site_id="+site_id if site_id else "")

            print(site_url)
            soup = html_parser(get(site_url).json()['data']['html'])
            anchors = soup.find_all(A)
            print('found', len(anchors), 'databases')
            for i in range(len(anchors)):
                x = anchors[i]
                if not x: break
                try:
                    memo[link.text][tag_to_str(x.contents[0]).strip()] = x.get('href')
                except Exception as ex:
                    print('error while parsing', link.text, 'databases:', ex)
        except Exception as ex:
            print('error while requesting', link.text, 'databases:', ex)

    return memo
```

```python
memo = go()
```

```python
from functools import reduce

reduce(lambda acc, x: acc + len(x[1].keys()), memo.items(), 0)
```
